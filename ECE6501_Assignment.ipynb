**Question 1**

In the cell below, plot the following activation functions for $x\in[-4, 4]$:

*   ReLU: $max(x, 0)$

```python
x = np.linspace(-4, 4, 50) # create a linear spacing from x = -4.0 to 4.0 with 50 steps
# create and plot the functions below

def ReLU(x):
    return np.maximum(x, 0)

ReLU = ReLU(x)
plt.plot(x, ReLU,color='blue', label="ReLU")
plt.grid(True)
plt.title('ReLU Function', fontsize=12)
plt.legend();
```

*   Leaky-ReLU: $max(0.1\cdot x, x)$


```python
x = np.linspace(-4, 4, 50) # create a linear spacing from x = -4.0 to 4.0 with 50 steps
# create and plot the functions below

def Leaky_ReLU(x):
    return np.maximum(0.1*x,x)

Leaky_ReLU = Leaky_ReLU(x)
plt.plot(x, Leaky_ReLU,color='blue', label="Leaky_ReLU")
plt.grid(True)
plt.title('Leaky ReLU Function', fontsize=12)
plt.legend();
```

*   Sigmoid: $\sigma(x) = 1/(1 + e^{-x})$

```python
x = np.linspace(-4, 4, 50) # create a linear spacing from x = -4.0 to 4.0 with 50 steps
# create and plot the functions below

Sigmoid = 1/(1+np.exp(-x))

plt.plot(x, Sigmoid,color='green', label="Sigmoid")
plt.grid(True)
plt.title('Sigmoid Function', fontsize=12)
plt.legend();
```

*   Hyperbolic Tangent: $\mathrm{tanh}(x) = (e^{x} - e^{-x})/(e^{x} + e^{-x})$

```python
x = np.linspace(-4, 4, 50) # create a linear spacing from x = -4.0 to 4.0 with 50 steps
# create and plot the functions below

Hyperbolic_Tangent = (np.exp(x) - np.exp(-x) )/(np.exp(x) + np.exp(-x))

plt.plot(x, Hyperbolic_Tangent,color='yellow', label="Hyperbolic_Tangent")
plt.grid(True)
plt.title('Hyperbolic Tangent Function', fontsize=12)
plt.legend();
```

*   SiLU: $x \cdot \sigma(x)$

```python
x = np.linspace(-4, 4, 50) # create a linear spacing from x = -4.0 to 4.0 with 50 steps
# create and plot the functions below

Sigmoid = 1/(1+np.exp(-x))
SiLU = x*Sigmoid
plt.plot(x, SiLU,color='black', label="SiLU")
plt.grid(True)
plt.title('Sigmoid Function', fontsize=12)
plt.legend();
```


*   GeLU: $x \cdot \frac{1}{2} \left(1 + \mathrm{erf}\left(\frac{x}{\sqrt{2}}\right)\right)$
*   tanh GELU: $x \cdot \frac{1}{2} \left(1 + \mathrm{tanh}\left(\frac{x}{\sqrt{2}}\right)\right)$

```python
x = np.linspace(-4, 4, 50) # create a linear spacing from x = -4.0 to 4.0 with 50 steps
# create and plot the functions below

length = len(x)

val1 = [math.erf(num/math.sqrt(2)) for num in x]
val2 = [math.tanh(num/math.sqrt(2)) for num in x]

GeLU = x*(1/2)*(np.array(1) + val1)
tanh_GeLU = x*(1/2)*(np.array(1) + val2)


plt.plot(x, GeLU,color='red', label="GeLU")
plt.legend()
plt.grid(True)
plt.title('GeLU Function', fontsize=12)
plt.show()

plt.plot(x, tanh_GeLU,color='red', linestyle='--',label="tanh GeLU")
plt.legend()
plt.grid(True)
plt.title('tanh GeLU Function', fontsize=12)
plt.show()
```

tanh is a common approximation as the error-function is computationally expensive to compute

```python
#plot GeLU and tanh GeLU in a same graph
plt.plot(x, GeLU,color='red', label="GeLU")
plt.plot(x, tanh_GeLU,color='red', linestyle='--',label="tanh GeLU")
plt.legend()
plt.grid(True)
plt.title('GeLU and tanh GeLU Function', fontsize=12)
plt.show()
```

**Question 2**

Answer to the following questions from the the plot you just created:

1. Which activation function is the least computationally expensive to compute?

   ReLU: $max(x, 0)$
   
   ReLU is the least computationally expensive to compute as it doesn't contain any multiplication, division  or exponentials operations. It simply gives an output, if x>0 otherwise it gives 0.

2. Are there better choices to ensure more stable training? What downfalls do you think it may have?

   Yes. ReLU causes vanishing gradient problem as well as since it is a linear function, it doesn't capture the non-linearity of some models.  Below functions ensure more stable trainings in such cases.

   * Leaky-ReLU: $max(0.1\cdot x, x)$ - This fuction addressing vanishing gradient problems better than the ReLU, as the gradient for negative inputs becomes smaller valaues in Leaky-ReLU. The downfall of this function is Leaky-ReLU is also a linear function and it doesn't address the non-linearities in some models.

   * Sigmoid: $\sigma(x) = 1/(1 + e^{-x})$ - This function addresses the non-linearity. The downfall of this function is because of the exponential operation, the complexity of the computations is high.

   * Hyperbolic Tangent: $\mathrm{tanh}(x) = (e^{x} - e^{-x})/(e^{x} + e^{-x})$ - This function addresses the non-linearity better. But because of the exponential operations, the complexity of the computations is high, and this function also causing for vanishing gradient problems.

   * SiLU: $x \cdot \sigma(x)$ - This function addresses the non-linearity better. But because of the exponential operations, the complexity of the computations is high.

   *   GeLU: $x \cdot \frac{1}{2} \left(1 + \mathrm{erf}\left(\frac{x}{\sqrt{2}}\right)\right)$ - This function addresses the non-linearity better. But because of the exponential operations, the complexity of the computations is high. This function may also causing for vanishing gradient problems.

   *   tanh GELU: $x \cdot \frac{1}{2} \left(1 + \mathrm{tanh}\left(\frac{x}{\sqrt{2}}\right)\right)$ - This function addresses the non-linearity better. But because of the tangent operations, the complexity of the computations is high. This function may also causing for vanishing gradient problems.



3. Are there any cases where you would not want to use either activation function?

   Yes. There are certain cases where the activation function not required.

   * Regression problems - In a regression problem, the final stage requires a prediction within a certain range. Having an activation function is not neccessary or irrelavant in this case as some of the negative values cancelled out with the activation function.

   * Special Cases - In some special neural networks such as generative models i.e. - predict the next word in a sequence, the input layer may process better ways without an activation function.
